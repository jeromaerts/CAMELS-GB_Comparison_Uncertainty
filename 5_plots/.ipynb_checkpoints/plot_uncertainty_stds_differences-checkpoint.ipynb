```python
import geopandas as gpd
import pandas as pd
import numpy as np

import cartopy
import cartopy.crs as ccrs
import seaborn as sns
import matplotlib
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt

from pathlib import Path
from glob import glob
```

```python
# Set Paths
ROOT = Path("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/")
AUXDATA = Path(f"{ROOT}/aux_data")
RESULTS = Path(f"{ROOT}/calibration_results/run_standard/wflow_sbm/")

# Set Config
basins_shp = f"{AUXDATA}/CAMELS-GB/data/CAMELS_GB_catchment_boundaries/CAMELS_GB_catchment_boundaries.shp"
results_csv = f"{ROOT}/results/wflow_sbm/calibration_overview_wflow.csv"
```

```python
# Read shapefile
gdf = gpd.read_file(basins_shp)
gdf = gdf.set_index('ID_STRING')
gdf.index = gdf.index.astype(int)

# Reproject shapefile
gdf = gdf.to_crs({'init':'epsg:4326'})
gdf['centroid'] = gdf.centroid.to_crs("EPSG:4326")
```

```python
# Load results dataframe
df_low = pd.read_csv("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/results/low_flow_uncertainty_overview_TEMP.csv", index_col='basin_id')
df_avg = pd.read_csv("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/results/mean_flow_uncertainty_overview_TEMP.csv", index_col='basin_id')
df_high = pd.read_csv("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/results/high_flow_uncertainty_overview_TEMP.csv", index_col='basin_id')


# Merge results and shapefile
gdf_low = gdf.merge(df_low, left_index=True, right_index=True, how="inner")
gdf_low = gdf_low.set_geometry("centroid")

gdf_avg = gdf.merge(df_avg, left_index=True, right_index=True, how="inner")
gdf_avg = gdf_avg.set_geometry("centroid")

gdf_high = gdf.merge(df_high, left_index=True, right_index=True, how="inner")
gdf_high = gdf_high.set_geometry("centroid")

# Remove rows that match observation nan rows
nan_rows = gdf_low[gdf_low['difference'].isna()]
gdf_low = gdf_low[~gdf_low.index.isin(nan_rows.index.to_list())]

nan_rows = gdf_avg[gdf_avg['difference'].isna()]
gdf_avg = gdf_avg[~gdf_avg.index.isin(nan_rows.index.to_list())]

nan_rows = gdf_high[gdf_high['difference'].isna()]
gdf_high = gdf_high[~gdf_high.index.isin(nan_rows.index.to_list())]

```

```python
fig, axes = plt.subplots(1, 3, figsize=(20, 20), subplot_kw={"projection": ccrs.PlateCarree()})

colormap = 'cool'

ax1 = gdf_low.plot(ax=axes[0],
                   column='largest_diff',
                   markersize=(gdf_low['difference']*30),
                   legend=True,
                   cmap=colormap)

ax1.add_feature(cartopy.feature.OCEAN)
ax1.add_feature(cartopy.feature.LAND, edgecolor='black')
ax1.add_feature(cartopy.feature.LAKES, edgecolor='black')
ax1.add_feature(cartopy.feature.RIVERS)


ax2 = gdf_avg.plot(ax=axes[1],
                   column='largest_diff',
                   markersize=(gdf_avg['difference']*30),
                   legend=True,
                   cmap=colormap)

ax2.add_feature(cartopy.feature.OCEAN)
ax2.add_feature(cartopy.feature.LAND, edgecolor='black')
ax2.add_feature(cartopy.feature.LAKES, edgecolor='black')
ax2.add_feature(cartopy.feature.RIVERS)


ax3 = gdf_high.plot(ax=axes[2],
                   column='largest_diff',
                   markersize=(gdf_high['difference']*30),
                   legend=True,
                   cmap=colormap)

ax3.add_feature(cartopy.feature.OCEAN)
ax3.add_feature(cartopy.feature.LAND, edgecolor='black')
ax3.add_feature(cartopy.feature.LAKES, edgecolor='black')
ax3.add_feature(cartopy.feature.RIVERS)


ax1.set_title('Low flow Uncertainty', size=20)
ax2.set_title('Average flow Uncertainty', size=20)
ax3.set_title('High flow Uncertainty', size=20)
# # ax4.set_title('KGE-NP Spearman R component', size=20)
# plt.tight_layout()
plt.savefig(f'{ROOT}/plots/draft_uncertainty_comparison.png')
```
