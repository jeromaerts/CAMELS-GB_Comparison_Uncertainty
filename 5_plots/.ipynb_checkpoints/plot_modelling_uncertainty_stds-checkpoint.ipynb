```python
import geopandas as gpd
import pandas as pd
import numpy as np

import cartopy
import cartopy.crs as ccrs
import seaborn as sns
import matplotlib
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt

from pathlib import Path
from glob import glob
```

```python
# Set Paths
ROOT = Path("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/")
AUXDATA = Path(f"{ROOT}/aux_data")
RESULTS = Path(f"{ROOT}/calibration_results/run_standard/wflow_sbm/")
MODELS = Path(f'{ROOT}/wflow/data/')
OBSDIR = Path(f"{AUXDATA}/CAMELS-GB/data/timeseries/")

# Set Config
basins_shp = f"{AUXDATA}/CAMELS-GB/data/CAMELS_GB_catchment_boundaries/CAMELS_GB_catchment_boundaries.shp"
```

```python
# Read shapefile
gdf = gpd.read_file(basins_shp)
gdf = gdf.set_index('ID_STRING')
gdf.index = gdf.index.astype(int)

# Reproject shapefile
gdf = gdf.to_crs({'init':'epsg:4326'})
gdf['centroid'] = gdf.centroid.to_crs("EPSG:4326")
```

```python
# Load results dataframe
df_low = pd.read_csv("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/results/low_flow_uncertainties_TEMP.csv", index_col='basin_id')
df_avg = pd.read_csv("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/results/average_flow_uncertainties_TEMP.csv", index_col='basin_id')
df_high = pd.read_csv("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/results/high_flow_uncertainties_TEMP.csv", index_col='basin_id')
```

```python
# Merge results and shapefile
gdf_low = gdf.merge(df_low, left_index=True, right_index=True, how="inner")
gdf_low = gdf_low.set_geometry("centroid")

gdf_avg = gdf.merge(df_avg, left_index=True, right_index=True, how="inner")
gdf_avg = gdf_avg.set_geometry("centroid")

gdf_high = gdf.merge(df_high, left_index=True, right_index=True, how="inner")
gdf_high = gdf_high.set_geometry("centroid")
```

```python
# sort values
sim_low = np.sort(df_low['model_difference_std'])
sim_avg = np.sort(df_avg['model_difference_std'])
sim_high = np.sort(df_high['model_difference_std'])

# Calculate rank ECDF
sim_y = np.arange(1, len(sim_low)+1) / len(sim_low)
```

# Modelling uncertainty

```python
fig, axes = plt.subplots(1, 3,figsize=(20, 20), subplot_kw={"projection": ccrs.PlateCarree()})

colormap1 = 'Reds'
colormap2 = 'coolwarm'
ax1 = gdf_low.plot(ax=axes[0], column='model_difference_std',norm=matplotlib.colors.LogNorm(vmin=gdf_low.model_difference_std.min(), vmax=gdf_low.model_difference_std.max()),
                        cmap=colormap1,
                        vmin=0,
                        vmax=5,
                        legend=True,
                        legend_kwds={'label': "Modelling Uncertainty standard deviation", 
                                     'orientation': "vertical",
                                      "shrink":.5,
                                    })

ax1.add_feature(cartopy.feature.OCEAN)
ax1.add_feature(cartopy.feature.LAND, edgecolor='black')
ax1.add_feature(cartopy.feature.LAKES, edgecolor='black')
ax1.add_feature(cartopy.feature.RIVERS)

ax2 = gdf_avg.plot(ax=axes[1], column='model_difference_std', norm=matplotlib.colors.LogNorm(vmin=gdf_low.model_difference_std.min(), vmax=gdf_low.model_difference_std.max()),
                        cmap=colormap1,
                        vmin=0,
                        vmax=5,
                        legend=True,
                        legend_kwds={'label': "Modelling Uncertainty standard deviation",
                                     'orientation': "vertical",
                                      "shrink":.5,
                                    })

ax2.add_feature(cartopy.feature.OCEAN)
ax2.add_feature(cartopy.feature.LAND, edgecolor='black')
ax2.add_feature(cartopy.feature.LAKES, edgecolor='black')
ax2.add_feature(cartopy.feature.RIVERS)


ax3 = gdf_high.plot(ax=axes[2], column='model_difference_std', norm=matplotlib.colors.LogNorm(vmin=gdf_low.model_difference_std.min(), vmax=gdf_low.model_difference_std.max()),
                        cmap=colormap1,
                        vmin=0,
                        vmax=5,
                        legend=True,
                        legend_kwds={'label': "Modelling Uncertainty standard deviation",
                                     'orientation': "vertical",
                                      "shrink":.5,
                                    })

ax3.add_feature(cartopy.feature.OCEAN)
ax3.add_feature(cartopy.feature.LAND, edgecolor='black')
ax3.add_feature(cartopy.feature.LAKES, edgecolor='black')
ax3.add_feature(cartopy.feature.RIVERS)


# sns.scatterplot(ax=axes[1,1], x=obs_low, y=obs_y, label='low flow', color='darkblue')
# sns.scatterplot(ax=axes[1,1], x=obs_avg, y=obs_y, label='average flow', color='darkgreen')
# sns.scatterplot(ax=axes[1,1], x=obs_high, y=obs_y, label='high flow', color='darkred')

# axes[1,1].set_title('Observation Uncertainty', size=20)
# axes[1,1].set_xlim(0,1)
# axes[1,1].set_xlabel('Standard Deviation')
# axes[1,1].set_ylabel('CDF')




ax1.set_title('Low flow modelling uncertainty std', size=20)
ax2.set_title('Average flow modelling uncertainty std', size=20)
ax3.set_title('High flow modelling uncertainty std', size=20)
# ax4.set_title('KGE-NP Spearman R component', size=20)
plt.tight_layout()
plt.savefig(f'{ROOT}/plots/draft_modelling_uncertainty.png')
```
