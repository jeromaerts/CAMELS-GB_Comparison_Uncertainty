```python
import geopandas as gpd
import pandas as pd
import numpy as np

import cartopy
import cartopy.crs as ccrs
import seaborn as sns
import matplotlib
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt

from pathlib import Path
from glob import glob
```

```python
# Set Paths
ROOT = Path("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/")
AUXDATA = Path(f"{ROOT}/aux_data")
RESULTS = Path(f"{ROOT}/calibration_results/run_standard/wflow_sbm/")
MODELS = Path(f'{ROOT}/wflow/data/')
OBSDIR = Path(f"{AUXDATA}/CAMELS-GB/data/timeseries/")

# Set Config
basins_shp = f"{AUXDATA}/CAMELS-GB/data/CAMELS_GB_catchment_boundaries/CAMELS_GB_catchment_boundaries.shp"
```

```python
# Read shapefile
gdf = gpd.read_file(basins_shp)
gdf = gdf.set_index('ID_STRING')
gdf.index = gdf.index.astype(int)

# Reproject shapefile
gdf = gdf.to_crs({'init':'epsg:4326'})
gdf['centroid'] = gdf.centroid.to_crs("EPSG:4326")
```

```python
# Load results dataframe
df_low = pd.read_csv("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/results/low_flow_uncertainties_TEMP.csv", index_col='basin_id')
df_avg = pd.read_csv("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/results/average_flow_uncertainties_TEMP.csv", index_col='basin_id')
df_high = pd.read_csv("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/results/high_flow_uncertainties_TEMP.csv", index_col='basin_id')
```

```python
# Merge results and shapefile
gdf_low = gdf.merge(df_low, left_index=True, right_index=True, how="inner")
gdf_low = gdf_low.set_geometry("centroid")

gdf_avg = gdf.merge(df_avg, left_index=True, right_index=True, how="inner")
gdf_avg = gdf_avg.set_geometry("centroid")

gdf_high = gdf.merge(df_high, left_index=True, right_index=True, how="inner")
gdf_high = gdf_high.set_geometry("centroid")
```

```python
# sort values
obs_low = np.sort(df_low['obs_difference_std'])
obs_low = obs_low[~np.isnan(obs_low)]
obs_avg = np.sort(df_avg['obs_difference_std'])
obs_avg = obs_avg[~np.isnan(obs_avg)]
obs_high = np.sort(df_high['obs_difference_std'])
obs_high = obs_high[~np.isnan(obs_high)]

# Calculate rank ECDF
obs_low_y = np.arange(1, len(obs_low)+1) / len(obs_low)
obs_avg_y = np.arange(1, len(obs_avg)+1) / len(obs_avg)
obs_high_y = np.arange(1, len(obs_high)+1) / len(obs_high)
```

```python
# sort values
sim_low = np.sort(df_low['model_difference_std'])
sim_avg = np.sort(df_avg['model_difference_std'])
sim_high = np.sort(df_high['model_difference_std'])

# Calculate rank ECDF
sim_low_y = np.arange(1, len(sim_low)+1) / len(sim_low)
sim_avg_y = np.arange(1, len(sim_avg)+1) / len(sim_avg)
sim_high_y = np.arange(1, len(sim_high)+1) / len(sim_high)
```

```python
fig, axes = plt.subplots(1,3, figsize=(20, 5))

sns.scatterplot(ax=axes[0], x=obs_low, y=obs_low_y, label='low flow obs', color='darkblue')
sns.scatterplot(ax=axes[1], x=obs_avg, y=obs_avg_y, label='average flow obs', color='darkgreen')
sns.scatterplot(ax=axes[2], x=obs_high, y=obs_high_y, label='high flow obs', color='darkred')

sns.scatterplot(ax=axes[0], x=sim_low, y=sim_low_y, label='low flow sim', color='tab:gray')
sns.scatterplot(ax=axes[1], x=sim_avg, y=sim_avg_y, label='average flow sim', color='tab:olive')
sns.scatterplot(ax=axes[2], x=sim_high, y=sim_high_y, label='high flow sim', color='tab:cyan')

axes[0].set_xlim(0,5)
axes[1].set_xlim(0,5)
axes[2].set_xlim(0,5)
axes[0].set_xlabel('Standard Deviation')
axes[1].set_xlabel('Standard Deviation')
axes[2].set_xlabel('Standard Deviation')
axes[0].set_ylabel('CDF')
axes[1].set_ylabel('CDF')
axes[2].set_ylabel('CDF')

axes[0].set_title('Low Flow std', size=20)
axes[1].set_title('Average Flow std', size=20)
axes[2].set_title('High Flow std', size=20)

plt.savefig(f'{ROOT}/plots/draft_uncertainty_cdfs.png', transparant=False)
```
