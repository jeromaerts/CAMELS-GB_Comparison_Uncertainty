# Plot objective function difference wflow_sbm Calibrated vs wflow_sbm Uncalibrated

## Load results

```python
def load_result_dataframe(wflow_calibrated_csv, wflow_uncalibrated_csv, pcrglob_csv, df_basin_ids):
    # Load dataframes
    df_wflow_calibrated = pd.read_csv(wflow_calibrated_csv, index_col='basin_id')
    df_wflow_uncalibrated = pd.read_csv(wflow_uncalibrated_csv, index_col='basin_id')
    df_pcrglob = pd.read_csv(pcrglob_csv, index_col='basin_id')
    
    # Combine dataframes
    df = df_basin_ids.join(df_wflow_calibrated, rsuffix='_wflow_calibrated')
    df = df.join(df_wflow_uncalibrated, rsuffix='_wflow_uncalibrated')
    df = df.join(df_pcrglob, rsuffix='_pcrglob')

    # Drop NaN values
    df = df[df['kge_np'].notna()]
    
    return df
```

```python
# Load Results
df_all =  load_result_dataframe(wflow_calibrated_csv, wflow_uncalibrated_csv, pcrglob_csv, df_basin_ids)
df_low =  load_result_dataframe(low_wflow_calibrated_csv, low_wflow_uncalibrated_csv, low_pcrglob_csv, df_basin_ids)
df_mean = load_result_dataframe(mean_wflow_calibrated_csv, mean_wflow_uncalibrated_csv, mean_pcrglob_csv, df_basin_ids)
df_high = load_result_dataframe(high_wflow_calibrated_csv, high_wflow_uncalibrated_csv, high_pcrglob_csv, df_basin_ids)
```

## Calculate difference in objective functions

```python
def calculate_objective_difference(df, objective_functions):
    for objective_function in objective_functions:
        df[f'diff_wflow_calibrated_vs_uncalibrated_{objective_function}' ] = df[f'{objective_function}'] - df[f'{objective_function}_wflow_uncalibrated']
        df[f'diff_wflow_calibrated_vs_pcrglob_{objective_function}' ] = df[f'{objective_function}'] - df[f'{objective_function}_pcrglob']
    return df
```

```python
# Calculate objective function differences
objective_functions = ['kge_np', 'kge_2009','kge_2012','nse']

df_all = calculate_objective_difference(df_all, objective_functions)
df_low = calculate_objective_difference(df_low, objective_functions)
df_mean = calculate_objective_difference(df_mean, objective_functions)
df_high = calculate_objective_difference(df_high, objective_functions)
```

## Read basin boundary shapefile

```python
# Read shapefile
gdf = gpd.read_file(basins_shp)
gdf = gdf.set_index('ID_STRING')
gdf.index = gdf.index.astype(int)

# Reproject shapefile
gdf = gdf.to_crs({'init':'epsg:4326'})
gdf['centroid'] = gdf.centroid.to_crs("EPSG:4326")
```

## Merge shapefile and dataframes

```python
# Merge results and shapefile
gdf_all = gdf.merge(df_all, left_index=True, right_index=True, how="inner")
gdf_all = gdf_all.set_geometry("centroid")

gdf_low = gdf.merge(df_low, left_index=True, right_index=True, how="inner")
gdf_low = gdf_low.set_geometry("centroid")

gdf_mean = gdf.merge(df_mean, left_index=True, right_index=True, how="inner")
gdf_mean = gdf_mean.set_geometry("centroid")

gdf_high = gdf.merge(df_high, left_index=True, right_index=True, how="inner")
gdf_high = gdf_high.set_geometry("centroid")
```
