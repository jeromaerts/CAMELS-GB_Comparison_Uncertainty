# Script that plots streamflow results

```python
import geopandas as gpd
import pandas as pd
import numpy as np

import cartopy
import cartopy.crs as ccrs
import seaborn as sns
import matplotlib
import matplotlib.gridspec as gridspec
import matplotlib.pyplot as plt

from pathlib import Path
from glob import glob
```

# Wflow calibration results

## Set Paths en Config

```python
# Set Paths
ROOT = Path("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/")
AUXDATA = Path(f"{ROOT}/aux_data")
RESULTS = Path(f"{ROOT}/calibration_results/run_standard/wflow_sbm/")

# Set Config
basins_shp = f"{AUXDATA}/CAMELS-GB/data/CAMELS_GB_catchment_boundaries/CAMELS_GB_catchment_boundaries.shp"
results_csv = f"{ROOT}/results/wflow_sbm/calibration_overview_wflow.csv"
```

## Read and Reproject basin shapefiles

```python
# Read shapefile
gdf = gpd.read_file(basins_shp)
gdf = gdf.set_index('ID_STRING')
gdf.index = gdf.index.astype(int)

# Reproject shapefile
gdf = gdf.to_crs({'init':'epsg:4326'})
gdf['centroid'] = gdf.centroid.to_crs("EPSG:4326")
gdf
```

## Load results and merge with shapefile

```python
# Load results dataframe
df = pd.read_csv(results_csv, index_col='basin_id')

# Merge results and shapefile
gdf_results = gdf.merge(df, left_index=True, right_index=True, how="inner")
gdf_results = gdf_results.set_geometry("centroid")
gdf_results
```

```python
gdf = gdf_results.loc[gdf_results.ksathorfrac > 1000]
```

```python
ksat_mins = []

for index, row in gdf_results.iterrows():
    df = pd.read_csv(f'/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/results/wflow_sbm/calibration/{index}_calibration_objective_functions.csv')
    mask = df.loc[df.kge_np >= df.kge_np.max() - 0.02]
    ksat = mask.ksathorfrac.min()
    ksat_mins.append(ksat)
```

```python
gdf_results['ksat_min'] = ksat_mins
df = pd.read_csv("/gpfs/work1/0/wtrcycle/users/jaerts/camels_uk/aux_data/CAMELS-GB/data/CAMELS_GB_hydrogeology_attributes.csv", index_col='gauge_id')
gdf_results = gdf_results.join(df)
gdf_results
```

## Plot calibration results

```python
# projection = ccrs.epsg(4326)
fig, ax = plt.subplots(figsize=(10, 15), subplot_kw={"projection": ccrs.PlateCarree()})
# colormap = 'bwr'
colormap = 'bwr_r'
# ax = plt.axes(projection=ccrs.CRS("WGS 84"))
# ax = plt.axes(projection=ccrs.PlateCarree())
# ax = plt.axes(projection=ccrs.OSGB())
# ax.set_global()
# ax.coastlines()

ax.coastlines(resolution='10m')

ax = gdf_results.plot(ax=ax, column='kge_np',
                        cmap=colormap,
                        vmin=-1.0,
                        vmax=1.0,
                        legend=True,
                        legend_kwds={'label': "KGE_NP value",
                                     'orientation': "vertical"})

ax.add_feature(cartopy.feature.OCEAN)
ax.add_feature(cartopy.feature.LAND, edgecolor='black')
ax.add_feature(cartopy.feature.LAKES, edgecolor='black')
ax.add_feature(cartopy.feature.RIVERS)

# plt.axis('off')
# plt.tight_layout()

ax.set_title('618 Basins, Calibration Period, KGE NP', size=20)
plt.savefig(f'{ROOT}/plots/calibration_results_wflow_sbm_UK.png', transparant=True)
```

## Plot calibration and evaluation wflow sbm CDFs

```python
# Set Config
results_calib_csv = f"{ROOT}/results/wflow_sbm/calibration_overview_wflow.csv"
results_eval_csv = f"{ROOT}/evaluation_results/calibration_overview_wflow.csv"

df_calib = pd.read_csv(results_calib_csv)
df_eval = pd.read_csv(results_eval_csv)
```

```python
# sort values
kge_np_calib = np.sort(df_calib['kge_np'])
kge_np_eval = np.sort(df_eval['kge_np'])

# Calculate rank ECDF
kge_y = np.arange(1, len(kge_np_calib)+1) / len(kge_np_calib)
```

```python
# Plot Figure
fig, ax = plt.subplots(figsize=(8, 8))


# KGE 2012
sns.scatterplot(ax=ax, x=kge_np_calib, y=kge_y, label='wflow_sbm calibration period', color='darkblue')
sns.scatterplot(ax=ax, x=kge_np_eval, y=kge_y, label='wflow_sbm evaluation period', color='darkgreen')

ax.set_title('CDF KGE_NP', size=20)
ax.set_xlim(-1,1)

plt.savefig(f'{ROOT}/plots/cdf_wflow_sbm_UK.png', transparant=True)
```

# Compare

```python
results_wflow_csv = f"{ROOT}/evaluation_results/calibration_overview_wflow.csv"
results_pcrglob_csv = f"{ROOT}/results/pcr-globwb/evaluation_overview_pcrglobwb_adjusted_loc.csv"
```

```python
df_wflow = pd.read_csv(results_wflow_csv, index_col='basin_id')
```

```python
df_pcrglob = pd.read_csv(results_pcrglob_csv, index_col='basin_id')
df_pcrglob
```

```python
df_eval = df_wflow.join(df_pcrglob, rsuffix='_pcrglob')
```

```python
df_eval['diff'] = df_eval['kge_np'] - df_eval['kge_np_pcrglob']
```

```python
# Read shapefile
gdf = gpd.read_file(basins_shp)
gdf = gdf.set_index('ID_STRING')
gdf.index = gdf.index.astype(int)

# Reproject shapefile
gdf = gdf.to_crs({'init':'epsg:4326'})
gdf['centroid'] = gdf.centroid.to_crs("EPSG:4326")

# Load results dataframe
df = pd.read_csv(results_adjusted_csv, index_col='basin_id')

# Merge results and shapefile
gdf_results = gdf.merge(df_eval, left_index=True, right_index=True, how="inner")
gdf_results = gdf_results.set_geometry("centroid")
len(gdf_results)
```

```python
fig, ax = plt.subplots(figsize=(10, 15), subplot_kw={"projection": ccrs.PlateCarree()})
colormap = 'bwr_r'

ax = gdf_results.plot(ax=ax, column='diff',
                        cmap=colormap,
                        vmin=-1.0,
                        vmax=1.0,
                        legend=True,
                        legend_kwds={'label': "KGE_NP value",
                                     'orientation': "vertical"})

ax.add_feature(cartopy.feature.OCEAN)
ax.add_feature(cartopy.feature.LAND, edgecolor='black')
ax.add_feature(cartopy.feature.LAKES, edgecolor='black')
ax.add_feature(cartopy.feature.RIVERS)

# plt.axis('off')
# plt.tight_layout()

ax.set_title('618 Basins diff kgeNP , wflow - PCR-GLOBWB ', size=20)
plt.savefig(f'{ROOT}/plots/comparison.png', transparant=True)
```
